```python
import pandas as pd
import os

input_folder = ' Others_Scripts/images_scripts_test/'

# Irei criar um dataframe vazio para armazenar os dados com as colunas que eu quero
dfs = []

for dim_file in os.listdir(input_folder):
    if dim_file.endswith('_stats.txt'):
    
        name = os.path.basename(dim_file)
        idate = name.split('_')[2].split('T')[0]

        df_virtual = pd.read_csv(os.path.join(input_folder, dim_file), sep='\t')
        df_virtual['date'] = pd.to_datetime(idate, format = '%Y%m%d')

        dfs.append(df_virtual)

# Concatena todos os dataframes em um nico dataframe
consolidated_df = pd.concat(dfs, ignore_index=True)

# Exibe o dataframe consolidado
display(consolidated_df)

consolidated_df.to_csv('Others_Scripts/images_scripts_test/results_final_bands.csv', index=False)


        
```
